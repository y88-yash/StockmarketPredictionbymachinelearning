```python
def predict(train,test,predictors, model) :
    model.fit(train[predictors],train["Traget"])
    pred=model.predict(test[predictors])
    pred=pd.Series(pred,index=test.index,name="prediction")
    combined=pd.concat([test["Traget"],pred],axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500,step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0],step):
        train= data.iloc[0:i].copy()
        test=data.iloc[i:(i+step)].copy()
        predictions = predict(train,test,predictors,model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions=backtest(tesla,model,predictors)
```

```python
predictions["prediction"].value_counts()
```

```python
 metrics.precision_score(predictions["Traget"],predictions["prediction"])
```

```python
predictions["Traget"].value_counts()/ predictions.shape[0]
```

```python
horizons=[2,5,60,250,1000]
new_predictors = []

for horizons in horizons:
    rolling_averages = tesla.rolling(horizons).mean()
    
    
    ratio_column = f"close_ratio_{horizons}"
    tesla[ratio_column] = tesla["Close"]/rolling_averages["Close"]
    
    trend_column = f"Trend_{horizons}"
    
    tesla[trend_column] = tesla.shift(1).rolling(horizons).sum()["Traget"]
    
    new_predictors += [ratio_column,trend_column]
    
```

```python
tesla = tesla.dropna()
```

```python
model = RandomForestClassifier(n_estimators=200,min_samples_split=50,random_state=1)

```

```python
def predict(train,test,predictors, model) :
    model.fit(train[predictors],train["Traget"])
    pred=model.predict_proba(test[predictors])[:,1]
    pred[pred >= .6]=1
    pred[pred < .6]=0
    pred=pd.Series(pred,index=test.index,name="prediction")
    combined=pd.concat([test["Traget"],pred],axis=1)
    return combined
```

```python
predictions = backtest(tesla,model,new_predictors)
```

```python
predictions["prediction"].value_counts()
```

```python
metrics.precision_score(predictions["Traget"],predictions["prediction"])
```
